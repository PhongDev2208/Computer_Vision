```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from PIL import Image

# c nh
image = Image.open('C:/Users/PhongIT/Pictures/animals.jpg')

# Chuyn sang nh xm
gray_image = image.convert('L')

image_array = np.array(gray_image)

# Chuyn sang nh nh phn
binary_image = 1*(image_array<200)

# Thc hin opening  loi b nhiu v ni cc i tng
opened_image = ndimage.binary_opening(binary_image, structure=np.ones((10,10)))

# Thc hin closing  loi b nhiu v ni cc i tng
closed_image = ndimage.binary_closing(binary_image, structure=np.ones((6,6)))

# nh nhn
labeled_image, num_features = ndimage.label(closed_image)

# To li subplot vi 5 dng v 1 ct
fig, axs = plt.subplots(5, 1, figsize=(50, 50))

plt.subplot(5,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title("nh xm")

plt.subplot(5,1,2)
plt.imshow(binary_image, cmap='gray')
plt.title("nh nh phn")
                                      
plt.subplot(5,1,3)
plt.imshow(opened_image, cmap='gray')
plt.title("nh sau opening")

plt.subplot(5,1,4)
plt.imshow(closed_image, cmap='gray')
plt.title("nh sau closing")

plt.subplot(5,1,5)
plt.imshow(labeled_image, cmap='gray')
plt.title(f" nh nhn ({num_features} i tng)")

plt.show();
```
