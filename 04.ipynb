```python
# Import lớp Image
from PIL import Image
from imgtools import *
from IPython.display import display # to display images
```

```python
# Đường dẫn ảnh
my_path = 'C:/Users/PhongIT/Pictures/dog.jfif'
```

```python
# Đọc ảnh 
img = load_image(my_path)
```

```python
display(img)
```

```python
# Định vị khu vực cắt
region = (70, 7, 215, 46)
# Cắt vùng ảnh
cropped_image = img.crop(region)
display(cropped_image)
```

```python
# Dán
# Xác định vị trí cần dán
paste_position = (142, 90)
img.paste(cropped_image, paste_position)
display(img)
```
