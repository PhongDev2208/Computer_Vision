```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Đường dẫn ảnh
my_path = 'C:/Users/PhongIT/Pictures/meomeme.jfif'
# Đọc ảnh
img = Image.open(my_path)
plt.imshow(img)
```

```python
# Chuyển sang ảnh xám
img_gray = img.convert('L')
plt.imshow(img_gray, cmap='gray')
```

```python
img_array = np.array(img_gray)
print(img_array)
```

```python
# Tạo một hình vẽ mới
plt.figure()
plt.hist(img_array.flatten(), bins=130)
plt.title('Biểu đồ Histogram')
plt.xlabel('Giá trị của pixel')
plt.ylabel('Số lượng')
plt.show()
```

```python
# Làm nổi bật đường biên
plt.figure()
plt.gray()
plt.contour(img_gray, origin="image")
plt.show()
```

```python
# Đường dẫn ảnh
my_path = 'C:/Users/PhongIT/Pictures/iphone15.jpeg'

# Đọc ảnh và hiển thị ảnh với trục tọa độ
img = Image.open(my_path).convert("L")

# Hiển thị ảnh trên trục tọa độ
plt.imshow(img)

# Tạo một ình vẽ mới
plt.figure()

# Đặt màu sắc là xám
plt.gray()

# Vẽ biểu đồ đường biên (contour) của ảnh
plt.contour(img, origin="image")

# Đảm bảo tỷ lệ trục x và trục y trên biểu đồ là bằng nhau
plt.axis('equal')

# Hiển thị ảnh
plt.show()
```
