```python
# Import lớp Image
from PIL import Image
```

```python
# Thư mục chứa hình ảnh
dir = 'C:/Users/PhongIT/Pictures'
```

```python
# Đường dẫn đến hình ảnh
image_path = dir +'/meomeme.jfif'
print(image_path)
```

```python
# Sử dụng Image.open() đọc ảnh
img01 = Image.open(image_path)
```

```python
# Hiển thị hình ảnh
display(img01)
```

```python
# In một số thông tin
# In định dạng ảnh
print("Định dang ảnh: ", img01.format)
```

```python

# In ra size
print("Kích thước hình ảnh: ", img01.size)
```

```python
# Lưu ảnh và chuyển đổi ảnh
dirOut = 'C:/Users/PhongIT/Documents/TGMT/images'
new_img_01_path = dirOut + '/meo.JPG'
new_img_png_01_path = dirOut + '/meo.PNG'
img01.save(new_img_01_path)
img01.save(new_img_png_01_path)
```

```python
# Đóng tập tin
img01.close()
```
